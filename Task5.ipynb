```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


data = pd.read_csv('US_Accidents_Dec21_updated.csv')
df = pd.DataFrame(data)
```

```python
df['Weather_Condition'].value_counts()
```

```python
cities = df['City'].unique()
len(cities)
```

```python
accidents_by_cities = df['City'].value_counts()
accidents_by_cities
```

```python
accidents_by_cities[:10]
```

```python
fig, ax = plt.subplots(figsize=(8,5))
accidents_by_cities[:10].plot(kind='bar')
ax.set(title = 'Top 10 cities By Number of Accidents',
       xlabel = 'Cities',
       ylabel = 'Accidents Count')
plt.show()
```

```python
accidents_severity = df.groupby('Severity').count()['ID']
accidents_severity
```

```python
fig, ax = plt.subplots(figsize=(8, 6), subplot_kw=dict(aspect="equal"))
label = [1,2,3,4]
plt.pie(accidents_severity, labels=label,
        autopct='%1.1f%%', pctdistance=0.85)
circle = plt.Circle( (0,0), 0.5, color='white')
p=plt.gcf()
p.gca().add_artist(circle)
ax.set_title("Accident by Severity",fontdict={'fontsize': 16})
plt.tight_layout()
plt.show()
```

```python

fig, ax = plt.subplots(figsize=(8, 5))  # Separate figsize for clarity

sns.histplot(data=df, x="Start_Time", bins=24)  # Use data argument for clarity

plt.xlabel("Start Time")
plt.ylabel("Number of Occurrences")  # Typo correction - Occurrences
plt.title('Accidents Count By Time of Day')

plt.show()
```

```python
Weather_Condition = df.groupby('Weather_Condition').count()['ID']
Weather_Condition
```

```python
fig, ax = plt.subplots(figsize=(8,5))
Weather_Condition.sort_values(ascending=False)[:20].plot(kind='bar')
ax.set(title = 'Weather Conditions at Time of Accident Occurence',
       xlabel = 'Weather',
       ylabel = 'Accidents Count')
plt.show()
```

```python
df.plot(kind='scatter', y='Start_Lat', x='Severity')
```
